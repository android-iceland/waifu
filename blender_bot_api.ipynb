```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("facebook/blenderbot-90M")

model = AutoModelForSeq2SeqLM.from_pretrained("facebook/blenderbot-90M")
def bot_reply(text):
  inputs = tokenizer([text], return_tensors="pt")
  reply_ids = model.generate(**inputs)
  reply=tokenizer.batch_decode(reply_ids, skip_special_tokens=True)[0]
  return reply

bot_reply("love you babe")  
```

```python
import gradio as gr
demo = gr.Interface(fn=bot_reply, inputs="text", outputs="text")

demo.launch(share=True)   
```
